```python
#задание 1

import pandas as pd
df = pd.read_csv('visit_log.csv',sep=';')
df.head()
```

```python
df.loc[df.traffic_source == 'yandex'or df.traffic_source == 'google', 'source_type'] = 'organic'
df.head()
```

```python
def sourse_type(row):
    if 'yandex' in row['traffic_source'] or 'google' in row['traffic_source']:
        return 'organic'
    elif 'paid' in row['traffic_source'] and 'Russia' in row['region'] or 'email' in row['traffic_source'] and 'Russia' in row['region']:
        return 'ad'
    elif 'paid' in row['traffic_source'] and 'Russia'not in row['region'] or 'email' in row['traffic_source'] and 'Russia'not in row['region']:
        return 'other'
    else:
        return row['traffic_source']
```

```python
df['source_type'] = df.apply(sourse_type, axis=1)
df.head(10)
```
