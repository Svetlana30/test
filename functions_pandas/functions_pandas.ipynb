```python
#задание 1
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def rating_movie(param):
    if param <= 2.0:
        return 'низкий рейтинг'
    if param <= 4.0:
        return 'средний рейтинг'
    if param == 4.5 or param == 5.0:
        return 'высокий рейтинг'
```

```python
ratings['class'] = ratings['rating'].apply(rating_movie)
ratings.sort_values('rating', ascending=False).head()
```

```python
#задание 2
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
df = pd.read_csv('keywords.csv')
df.head()
```

```python
def search (data):
    for key, values in geo_data.items():
        for value in values:
            if value in data:
                return key
    return 'undefined'
```

```python
df['region'] = df['keyword'].apply(search)
df[df.region != 'undefined'].head(20)
```
